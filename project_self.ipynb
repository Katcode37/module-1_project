```python
import numpy as np
import pandas as pd
import statistics as st
from scipy.stats import zscore
```

```python
exam_results = pd.read_csv('exam_dataset.csv', sep=';')
exam_results.head()
exam_results.tail()

```

```python
#exam_results.info()
#exam_results.isnull().sum()
#"Type of content" has missing values, but it is not a mistake. 
exam_results["Type of content"] = exam_results["Type of content"].fillna("No")
exam_results = exam_results.dropna()
exam_results.isnull().sum()
exam_results.info()
#noticed that there are 49 entries, but 45 rows. Index needs resetting.
exam_results.reset_index(drop=True) 
#Now we droped all the rows with missing values, which probably happened due to the fact that students did not answer some questions. We can also see that all the columns have the right type of content, so we can start our analysis.
                                                                    
```

```python
#Next huge problem - different tests are scored differently, but we need to make them comparable. 
# Z-score is a way to standardize scores on the same scale by converting them to a common scale with a mean of 0 and a standard deviation of 1. 
score_columns = [
    "Overall exam score",
    "Listening score",
    "Reading score",
    "Speaking score",
    "Writing Score"
]
for column in score_columns:
    exam_results[column + "_z"] = exam_results.groupby("Exam type")[column].transform(zscore)
exam_results  
```

```python
#Now we can compare the score of the students with no media exposure to the students with media exposure. 

yes = exam_results[exam_results["Informal English exposure"] == "Yes"]["Overall exam score_z"]
no = exam_results[exam_results["Informal English exposure"] == "No"]["Overall exam score_z"]

#Independent Samples t-test?
#Null hypothesis
#Students with exposure and without exposure have the same mean score.
#Alternative hypothesis
#Students with exposure have different (or higher) scores.
from scipy.stats import ttest_ind
t_stat, p_value = ttest_ind(yes, no)
print(t_stat)
print(p_value)

if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant difference in scores between students with and without media exposure.")
    if yes.mean() > no.mean():
        print("Students with media exposure have higher scores on average.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in scores between students with and without media exposure.")


```

```python
#Now when we know there is a significant difference, we can check the correlation between the scores and the media exposure, to see if there is a positive correlation between them: 
#the more hours of media exposure, the higher the score.
#We can use Pearson correlation coefficient for this.
```

```python
from scipy.stats import pearsonr
r, p_value = pearsonr(exam_results["Hours per week"], exam_results["Overall exam score_z"])
print(p_value)

if p_value < 0.05:
    print("There is a significant correlation between hours of media exposure and overall exam scores.")
    if r > 0:
        print("The correlation is positive: as hours of media exposure increase, overall exam scores tend to increase.")
    else:
        print("The correlation is negative: as hours of media exposure increase, overall exam scores tend to decrease.")
else:
    print("There is no significant correlation between hours of media exposure and overall exam scores.")
```

```python
#There is a clear evidence that the increase in hours of media exposure is associated with an increase in overall exam scores.
#Now it seems reasonable to find the threshold of hours of media exposure that is associated with a significant increase in scores. We can use a simple linear regression for this.
unique_hours = sorted(exam_results["Hours per week"].unique())
unique_hours
#Now we got out unique values and we can build our "steps" to find a threshold.
```

```python
def exposure_group(hours):
    if hours <= 2:
        return "Low"
    elif hours <= 7:
        return "Medium"
    else:
        return "High"

exam_results["Exposure group"] = exam_results["Hours per week"].apply(exposure_group)
exam_results.head()
exam_results["Exposure group"].value_counts()
exam_results.groupby("Exposure group")["Overall exam score_z"].mean()

#To escape multiple comparisons problem - ANOVA test 
from scipy.stats import f_oneway
low = exam_results[exam_results["Exposure group"] == "Low"]["Overall exam score_z"]
med = exam_results[exam_results["Exposure group"] == "Medium"]["Overall exam score_z"]
high = exam_results[exam_results["Exposure group"] == "High"]["Overall exam score_z"]
f_stat, p_value = f_oneway(low, med, high)
if p_value < 0.05:
    print("There is a significant difference in overall exam scores among the different exposure groups.")
else:
    print("There is no significant difference in overall exam scores among the different exposure groups.")

#there is a significant difference between the groups, but we need to find out which groups are different from each other
from statsmodels.stats.multicomp import pairwise_tukeyhsd
tukey = pairwise_tukeyhsd(endog=exam_results["Overall exam score_z"], groups=exam_results["Exposure group"], alpha=0.05)
print(tukey)



```

```python
#High vs Medium = 1.2982 of mean difference, which is 1.3 standard deviations.
#Important to understand what does it mean in terms of actual points if we want target audience to understand the result.
one_exam_std = exam_results.groupby("Exam type")["Overall exam score"].std()
understandable_exact_exam_points = one_exam_std * 1.3
print(understandable_exact_exam_points)
```

The results of the Tukey HSD test show that there is a significant difference in overall exam scores between the Medium and High exposure groups. 
However, there is no significant difference between the Low and Medium exposure groups. 
This suggests that students with high media exposure have significantly higher overall exam scores compared to those with low or medium media exposure.
Precicely, 
1 hour a day makes a huge difference in exam scores.
Moreove, this 1 hour a day allows students to increase their result by:
FCE      16 points up
IELTS     1 full band up 
TOEFL    32 points up

```python
#Okey obviously the results are astonishing. But which particular skills increase exactly? Does media exposure increase all the skills equally, or some skills more than others?

skills = [
    "Listening score_z",
    "Reading score_z",
    "Writing Score_z",
    "Speaking score_z"
]

for skill in skills:
    low = exam_results[exam_results["Exposure group"]=="Low"][skill]
    med = exam_results[exam_results["Exposure group"]=="Medium"][skill]
    high = exam_results[exam_results["Exposure group"]=="High"][skill]
    F, p_value = f_oneway(low, med, high)
    print(f"ANOVA shows for {skill}: F = {round(F,2)}, p = {round(p_value,4)}")

    tukey = pairwise_tukeyhsd(
        endog = exam_results[skill],
        groups = exam_results["Exposure group"],
        alpha = 0.05
    )
    print(f"Tuckey HSD for {skill}:")
    print(tukey.summary())
#According to Tuckey 
#Writing  → 1.16 SD
#Speaking → 1.08 SD
#Listening → 0.87 SD
```

Media exposure is associated with higher writing scores, as well as speaking and listening scores (in exactly that order).
It could be suggested that frequent listening aids internalizing structures and increases motivations; students might mimic grammar and language structures, memorizing them better.

```python
exam_results["Type of content"].value_counts()

def content_type(string):
    if string == "YouTube":
        return "YB"
    elif string == "Short vids":
        return "Shorts"
    elif string == "Series" or string == "Films":
        return "Series/Films"
    else:
        return "Nothing"
    
exam_results["Content groups"] = exam_results["Type of content"].apply(content_type)

nothing = exam_results[exam_results["Content groups"]=="Nothing"]["Overall exam score_z"]
shorts = exam_results[exam_results["Content groups"]=="Shorts"]["Overall exam score_z"]
yb = exam_results[exam_results["Content groups"]=="YB"]["Overall exam score_z"]
series_films = exam_results[exam_results["Content groups"]=="Series/Films"]["Overall exam score_z"]
F, p_value = f_oneway(nothing, shorts, yb, series_films)

print(f"F = {round(F,2)}")
print(f"p = {p_value}") #p_value tells us content type is important, Tuckey to find out which one.

tukey = pairwise_tukeyhsd(
    endog=exam_results["Overall exam score_z"],
    groups=exam_results["Content groups"],
    alpha=0.05
)

print(tukey)
 

```

Short-form video exposure is associated with significantly higher exam scores compared to none.
Other content types also show higher scores overall than no exposure, but these differences were not as drastic as shorts effect demonstrated. 

```python
#Do students with many years learning English but low media exposure still score worse?
 

exam_results["High_years"] = exam_results["Years learning English"] >= 10

high_years_low_exposure = exam_results[
    (exam_results["High_years"] == True) &
    (exam_results["Exposure group"] == "Low")
]["Overall exam score_z"]

high_exposure_less_years = exam_results[
    (exam_results["Exposure group"] == "High") & (exam_results["Years learning English"] < 10)
]["Overall exam score_z"]

t_stat, p_value = ttest_ind(high_years_low_exposure, high_exposure_less_years)
print(f"t = {round(t_stat,2)}")
print(f"p = {p_value}")

if p_value < 0.05:
    print("There is a significant difference in overall exam scores between students with many years of learning English but low media exposure and students with high media exposure.")
    if high_years_low_exposure.mean() > high_exposure_less_years.mean():
        print("Students with many years of learning English but low media exposure have higher scores on average.")
    else:
        print("Students with high media exposure have higher scores on average even if they have fewer years of learning English.")

```

Those engaging in video content more then 1 hour daily - but learned english for a shorter period of time then 10 years -  perform better then even those who spent more then 10 years learning English!

```python
# Do student preparing for the exam with media exposure score better than students spending more time actually preparing fot the exam?

exam_results["High_preparation"] = exam_results["Months preparing for exam"] >= 6
high_preparation = exam_results[(exam_results["High_preparation"] == True) & (exam_results["Exposure group"] == "Low")]["Overall exam score_z"]
high_exposure_less_months_prep = exam_results[
    (exam_results["Exposure group"] == "High") & (exam_results["Months preparing for exam"] < 6)
]["Overall exam score_z"]

t_stat, p_value = ttest_ind(high_preparation, high_exposure_less_months_prep)
print(f"t = {round(t_stat,2)}")
print(f"p = {p_value}")
if p_value < 0.05:
    print("There is a significant difference in overall exam scores between students who prepared for the exam for 6 or more months and students with high media exposure.")
    if high_preparation.mean() > high_exposure_less_months_prep.mean():
        print("Students who prepared for the exam for 6 or more months have higher scores on average.")
    else:
        print("Students with high media exposure have higher scores on average.")
```

Those engaging in video content more then 1 hour daily perform better then even those who spent more time preparing.
